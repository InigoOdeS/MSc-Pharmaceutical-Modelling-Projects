```python
def table(f):
    d = open(f,"r")
    print('Name','\t\t','%GC','\t\t\t','Avequal')
    n = 0
    j = list()
    k = list()
    l = list()
    for line in d:
        line = line.strip()
        if n % 4 == 0:
            line = list(line)
            line[0] = ""
            line = "".join(line)
            j.append(line)
        if n % 4 == 1:
            line1 = list(line)
            A = line1.count("A")
            C = line1.count("C")
            T = line1.count("T")
            G = line1.count("G")
            q = ((C+G)/(A+C+T+G))*100
            k.append(q)
        if n % 4 == 3:
            line2 = list(line)
            qual = list()
            for x in line2:
                qual.append(ord(x))
            ss = sum(qual)
            qaver = ss/len(line2)
            l.append(qaver)
        n += 1
            
    for a in range(len(j)):
        print("".join([j[a],'\t', str(k[a]),'\t', str(l[a])]))
    
```

```python
print(table("3FB221_project1.fq"))
```

```python
import re
def fastaconvert(f):
    d = open(f,"r")
    n = 0
    for line in d:
        line = line.strip()
        if n % 4 == 0:
            line = list(line)
            line[0] = ">"
            line = "".join(line)
            print(line)
        if n % 4 == 1:
            for v in re.findall('.{1,30}', line):
                print(v)
        n += 1
```

```python
print(fastaconvert("3FB221_project1.fq"))
```

```python
def putspace(line): #used to get a list where its elements are 3 letters elements 
    N=1
    Pist = [] 
    for i in range(len(line)) :
        if N % 3 == 0:
            Pist.append(line[i-2] +line[i-1]+line[i])
        N += 1
    return(Pist) 

def conversion(prot):#this function converts the list of 1-letters to a sequence which corresponds to the protein
    t = 0
    for i in range(len(prot)) :
        if prot[i] == 'M' :
            j=i
            protein='' 
            while j < len(prot) and prot[j] != '_' and j > t:
                protein += prot[j] 
                j += 1
            t = j
            print(protein)

def convertprotein(f):
    d = open(f,"r")
    n = 0
    for line in d:
        line = line.strip()
        if n % 4 == 0:
            line = list(line)
            line[0] = ">"
            line = "".join(line)
            print(line)
        if n % 4 == 1:
            digitslist = putspace(line)
            codontable = {
            'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M', 
            'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T', 
            'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K', 
            'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',                  
            'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L', 
            'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P', 
            'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q', 
            'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R', 
            'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V', 
            'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A', 
            'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E', 
            'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G', 
            'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S', 
            'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L', 
            'TAC':'Y', 'TAT':'Y', 'TAA':'_', 'TAG':'_', 
            'TGC':'C', 'TGT':'C', 'TGA':'_', 'TGG':'W'
            }
            prot=[] 
            for j in digitslist:
                prot.append(codontable.get(j))
            print(conversion(prot)) 
            
        n += 1
```
